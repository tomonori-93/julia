# Lilly (1966) の Ekman 層不安定固有値解析 (Chebyshev 多項式展開)
[Lilly (1966)](https://doi.org/10.1175/1520-0469(1966)023<0481:OTIOEB>2.0.CO;2) による Ekman 層における Parallel instability の固有値解析 (2 変数の Orr-Sommerfeld 方程式) を Chebyshev 多項式展開で実施する.

ただし, 切断波数$N$が大きい場合, 4 階微分項の評価が $O(N^7)$となり, 固有値計算時の丸め誤差が無視できなくなる可能性がある ([Dongarra et al. 1996](https://doi.org/10.1016/S0168-9274(96)00049-9)). そのため, 中間変数$\chi $を導入して 2 階微分のみで計算する ([Hibino et al. 2012](https://doi.org/10.2151/jmsj.2012-210)). 

\begin{equation}
\ddot{\varphi} -\alpha ^2\varphi \equiv \chi , \label{eq:B1-01}
\end{equation}
となる中間変数$\chi $を導入する. まず, チェビシェフ展開するために, 微分方程式の定義域を$0\leq z'\leq z'_{\mathrm{top}}\rightarrow \infty $から, $-1\leq Z\leq 1$に座標変換する:
$$Z\equiv f(z')=G^{-1}\left(z'-G \right) ,\quad G\equiv \dfrac{z'_{\mathrm{top}}}{2} $$
すると, (\ref{eq:B1-01}) 式は
\begin{equation}
\ddot{\varphi} -G^2\left(\alpha ^2\varphi +\chi \right) =0. \label{eq:B1-02}
\end{equation}
Lilly (1966) の (25), (26) 式は
\begin{equation}
\ddot{\chi} -G^2\alpha ^2\chi -i\alpha R\left[G^2(V-c)\chi -\ddot{V} \varphi \right] +2G\dot{\mu} =0. \label{eq:B1-03}
\end{equation}
\begin{equation}
\ddot{\mu} -G^2\alpha ^2\mu -i\alpha R\left[G^2(V-c)\mu +G\dot{U} \varphi \right] -2G\dot{\varphi} =0. \label{eq:B1-04}
\end{equation}
Orszag (1971) の (13) 式にしたがって, 各変数を
\begin{equation}
\left[\begin{array}{c}
\varphi (Z) \\
\chi (Z) \\
\mu (Z) \\
U (Z) \\
V (Z) \\
\end{array} \right] =\sum^{\infty}_{n=0}{\left[\begin{array}{c}
\hat{\varphi} _n \\
\hat{\chi} _n \\
\hat{\mu} _n \\
\hat{U} _n \\
\hat{V} _n \\
\end{array} \right] T_n(Z)} , \label{eq:B1-05} 
\end{equation}
と展開する. (\ref{eq:B1-02})-(\ref{eq:B1-04}) 式のチェビシェフ展開はそれぞれ
\begin{align}
&\; \sum^{\infty}_{n=0}{\left\{\hat{\varphi}^{(2)}_n-G^2\left(\alpha ^2\hat{\varphi}_n+\hat{\chi}_n \right) \right\} T_n} =0, \label{eq:B1-07} \\
&\; \sum^{\infty}_{n=0}{\left\{\hat{\chi}^{(2)}_n-G^2\alpha ^2\hat{\chi}_n-\dfrac{i\alpha R}{2} \left[G^2e_{V,\chi ,n}-f^{(2)}_{V,\varphi ,n} \right] +i\alpha RcG^2\hat{\chi}_n+2G\hat{\mu}^{(1)}_n \right\} T_n} =0, \label{eq:B1-08} \\
&\; \sum^{\infty}_{n=0}{\left\{\hat{\mu}^{(2)}_n-G^2\alpha ^2\hat{\mu}_n-\dfrac{i\alpha R}{2} \left[G^2e_{V,\mu ,n}+Gf^{(1)}_{U,\varphi ,n} \right] +i\alpha RcG^2\hat{\mu}_n-2G\hat{\varphi}^{(1)}_n \right\} T_n} =0, \label{eq:B1-09} 
\end{align}
ただし, Orszag (1971) の付録 (A7) より, 非線形項は
\begin{equation}
\begin{split}
e_{V,\chi ,n}\equiv &\; \dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{V}_mc_{|n-m|}\hat{\chi}_{|n-m|}}, \\
e_{V,\mu ,n}\equiv &\; \dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{V}_mc_{|n-m|}\hat{\mu}_{|n-m|}}, \\
f^{(2)}_{V,\varphi ,n} \equiv &\; \dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{\varphi}_mc_{|n-m|}\hat{V}^{(2)}_{|n-m|}} =\dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{\varphi}_m\sum^{\infty}_{\substack{p=|n-m|+2\\ p\equiv |n-m|\; (\mathrm{mod} \; 2)}}{p(p^2-(n-m)^2)\hat{V}_p}} , \\
f^{(1)}_{U,\varphi ,n} \equiv &\; \dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{\varphi}_mc_{|n-m|}\hat{U}^{(1)}_{|n-m|}} =\dfrac{2}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{\varphi}_m\sum^{\infty}_{\substack{p=|n-m|+1\\ p+|n-m|\equiv 1\; (\mathrm{mod} \; 2)}}{p\hat{U}_p}} .
\end{split}
 \label{eq:B1-10}
\end{equation}
境界条件について, Lilly (1966) の (27) 式を$Z=-1$, (28) 式を$Z=1$で成り立つようにすると,
\begin{equation}
(27):\quad \sum^{N}_{p=0}{(-1)^p\hat{\varphi}_p} =0,\quad \sum^{N}_{p=1}{p^2(-1)^{p-1}\hat{\varphi}_p} =0,\quad \sum^{N}_{p=0}{(-1)^p\hat{\mu}_p} =0,\quad (Z=-1), \label{eq:B1-19}
\end{equation}
\begin{equation}
(28):\quad \sum^{N}_{p=0}{(+1)^p\hat{\varphi}_p} =0,\quad \sum^{N}_{p=0}{(+1)^p\hat{\chi}_p} =0,\quad \sum^{N}_{p=1}{p^2(+1)^{p-1}\hat{\mu}_p} =0,\quad (Z=1). \label{eq:B1-20}
\end{equation}
(\ref{eq:B1-20}) の第 2 式は$Z=1$では$\varphi = \ddot{\varphi} =0$なので, (\ref{eq:B1-02}) 式から$\chi =0$が成り立つことを利用している. 

(\ref{eq:B1-16})-(\ref{eq:B1-18}) 式は$T_n$が$0\leq n\leq N-2$のそれぞれについて$3(N-1)$個の独立な方程式となる. これに加えて境界条件が 6 つの追加の関係を各係数$\hat{\varphi}_n,\; \hat{\chi}_n,\; \hat{\mu}_n$について与えるので, $3(N+1)$個の未知変数について, $3(N+1)$個の方程式が組み立てられる. 今, (\ref{eq:B1-16})-(\ref{eq:B1-18}) 式および境界条件 (\ref{eq:B1-19})-(\ref{eq:B1-20}) 式から
\begin{equation}
A\textbf{x} =cB\textbf{x} , \label{eq:B1-21a}
\end{equation}
\begin{equation}
\textbf{x} \equiv [\hat{\varphi}_0,\cdots ,\hat{\varphi}_N,\hat{\chi}_0,\cdots ,\hat{\chi}_N,\hat{\mu}_0,\cdots ,\hat{\mu}_N]^T \label{eq:B1-21}
\end{equation}
という形の行列方程式が構築できる. $\textbf{x} $は未知変数で構築される$3(N+1)$次元の列ベクトル, $A,\; B$は$3(N+1)$次正方行列である. ここで, $A,B$はそれぞれ
\begin{equation}
A\equiv \left(\begin{array}{ccc}
A_{\ref{eq:B1-16},\hat{\varphi}} & A_{\ref{eq:B1-16},\hat{\chi}} & A_{\ref{eq:B1-16},\hat{\mu}} \\
\mathrm{BC1}   & \textbf{0}^T  & \textbf{0}^T \\
\mathrm{BC2}   & \textbf{0}^T  & \textbf{0}^T \\
A_{\ref{eq:B1-17},\hat{\varphi}} & A_{\ref{eq:B1-17},\hat{\chi}} & A_{\ref{eq:B1-17},\hat{\mu}} \\
\mathrm{BC3}   & \textbf{0}^T  & \textbf{0}^T \\
\textbf{0}^T   & \mathrm{BC4}  & \textbf{0}^T \\
A_{\ref{eq:B1-18},\hat{\varphi}} & A_{\ref{eq:B1-18},\hat{\chi}} & A_{\ref{eq:B1-18},\hat{\mu}} \\
\textbf{0}^T   & \textbf{0}^T  & \mathrm{BC5} \\
\textbf{0}^T   & \textbf{0}^T  & \mathrm{BC6} \\
\end{array} \right) ,\quad B\equiv \left(\begin{array}{ccc}
O          &      O     &      O     \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T \\
O          & B_{\ref{eq:B1-17},\chi} & O \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T \\
O          & O & B_{\ref{eq:B1-18},\mu} \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T \\
\end{array} \right) , \label{eq:B1-22}
\end{equation}
ただし, 区分行列$A_{j,k}, B_{j,k}$は支配方程式 ($j$) において, 変数$k$の係数を表す$N-1$行$N+1$列の行列, $O$は$N-1$行$N+1$列のゼロ行列, $\textbf{0}$は$N+1$次のゼロベクトルとする. BC1-BC6 は 6 つの境界条件から得られる$\hat{\varphi}_n,\; \hat{\chi}_n,\; \hat{\mu}_n$に対する関係式である. 


```julia
module Hibino2012_solver

using OffsetArrays  # 配列要素番号を任意開始にするモジュール (行列計算時は 1 始まりの通常 matrix に copy する)
using LinearAlgebra
using FFTW

export full_EQ_set, OSE_set, basic_flow_P2T

function full_EQ_set( α_in, ε_in, Re_in, z_top, N_in )
# Solve the complete set of the equations (2.25)-(2.28) in Lilly (1966)
    #-- arguments input:
    # α_in: wavenumber for y-direction [non-dim,Vector{Float64}]
    # ε_in: flow angle for Vg [degree,Vector{Float64}]
    # Re_in: Reinolds number [non-dim,Vector{Float64}]
    # Vb_in: Background flow spectra for V component [non-dim,Vector{Float64}]
    # Ub_in: Background flow spectra for U component [non-dim,Vector{Float64}]
    # z_top: Reinolds number [non-dim,Vector{Float64}]
    # N_in: truncation wavenumber in vertical [1,Int64]
    #-- arguments output:
    # c_out: wave speed with the maximum in imaginary part [non-dim,Array{Complex{Float64},size(α),size(ε),size(Re)}]

    # set parameters
    d2r = π / 180.0  # degree to radian
    na = size(α_in)[1]
    ne = size(ε_in)[1]
    nr = size(Re_in)[1]
    N = N_in
    Ncal = N + 1
    Ntot = 3 * (N + 1)
    Gf = 0.5 * z_top
    G2 = Gf^2

    # initialize variables and matrix
    c_out = reshape(zeros(Complex{Float64},na,ne,nr),na,ne,nr)

    # initialize working arrays
    # A: non-lambda coefficients, B: lambda coefficients
    A_J = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    B_J = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    ### --- From here, supposing matrix element as row, col as in mathematical notation
    Acal = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    Bcal = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    A_φφ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for v in φ Chebyshev
    A_φχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in φ Chebyshev
    A_χφ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for φ in χ Chebyshev
    A_χχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in χ Chebyshev
    A_χμ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for μ in χ Chebyshev
    A_μφ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for φ in μ Chebyshev
    A_μχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in μ Chebyshev
    A_μμ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for μ in μ Chebyshev
    B_χχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in χ Chebyshev
    B_μμ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for μ in μ Chebyshev
    ### --- From here, supposing matrix element as col, row as in julia notation (only using these arrays for matrix calc)
    Vb = OffsetArray(zeros(Float64,Ncal),0:N)  # A vector composed of Chebyshev coefficients for the basic state flow of V component
    Ub = OffsetArray(zeros(Float64,Ncal),0:N)  # A vector composed of Chebyshev coefficients for the basic state flow of V component
    Vb_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of V component in the physical space
    Ub_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of U component in the physical space
    c = OffsetArray(zeros(Float64,Ncal),0:N)  # factorized vector (c_0=2, c_n=1, (n>0))
    #-- For specifying the basic flows in physical space z_p (0 <= z_p <= z_top)
    θ = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in vertical coordinate (-π <= θ <= 0)
    Z = OffsetArray(zeros(Float64,Ncal),0:N)  # non-dimensional vertical coordinate (-1 <= Z <= 1)
    z_p = OffsetArray(zeros(Float64,Ncal),0:N)  # dimensional vertical coordinate (0 <= z_p <= z_top)

    # setting the factorized vector
    c[0] = 2.0
    c[1:N] .= 1.0

#    # setting basic flow spectra
#    Vb[0:N] .= Vb_in[1:Ncal]
#    Ub[0:N] .= Ub_in[1:Ncal]
    
    # set basic flow profiles in vertical
    θ[0:N] = [π * (j-N) / N for j in 0:N]  # non-dimensional vertical coordinate (-1<=Z<=1)
    Z[0:N] .= cos.(θ[0:N])
    z_p[0:N] .= Gf .* (Z[0:N] .+ 1.0)

    # setting maximum even and odd numbers from N
    Nemax = 0
    Nomax = 1
    for i in 0:2:N
        Nemax = i
    end
    for i in 1:2:N
        Nomax = i
    end
    #println("Nemax = ", Nemax, ", Nomax = ", Nomax)

    #-- Start loops for varying parameters (α, ε, Re)
    for k in 1:nr
        for j in 1:ne
            for i in 1:na
                α = α_in[i]
                ε = ε_in[j] * d2r
                Re = Re_in[k]

                # initialize each matrices
                Acal = zeros(Complex{Float64},Ntot,Ntot)
                Bcal = zeros(Complex{Float64},Ntot,Ntot)
                A_φφ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_φχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χφ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χμ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_μφ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_μχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_μμ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                B_χχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                B_μμ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]

                # set grid spacing and mean-wind profiles in vertical
                Ub_phys[0:N] .= cos(ε) .- exp.(-z_p[0:N]) .* cos.(z_p[0:N] .+ ε)
                Vb_phys[0:N] .= -sin(ε) .+ exp.(-z_p[0:N]) .* sin.(z_p[0:N] .+ ε)

                Vb = basic_flow_P2T( N, z_top, Vb_phys[0:N] )
                Ub = basic_flow_P2T( N, z_top, Ub_phys[0:N] )

                # setting matrices
                # 支配方程式から構築される部分 (チェビシェフ次数 0<=n<=N-2 の範囲)
                # この段階では境界条件で成り立つ係数の関係は考慮しない
                for n in 0:N-2
                    n2 = n^2
                    α2 = α^2
                    
                    # In v rows, a_{φ,n} and a_{χ,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_φφ[n,p] = p * (p2 - n2)
                    end
                    
                    A_φφ[n,n] = - G2 * c[n] * α2
                    A_φχ[n,n] = - G2 * c[n]
                    
                    # In χ rows, a_{φ,n}, a_{χ,n}, and a_{μ,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_χχ[n,p] = p * (p2 - n2)
                    end
                    
                    A_χχ[n,n] = - G2 * c[n] * α2
                    
                    for p in -N:N
                        abs_n_p = abs(n - p)
                        abs_p = abs(p)
                        if abs_n_p <= N
                            A_χχ[n,abs_p] = A_χχ[n,abs_p] + complex(0.0, - 0.5 * G2 * α * Re * c[abs_p] * c[abs_n_p] * Vb[abs_n_p])
                        end
                        
                        if abs_n_p+2 <= N
                            acoef = 0.0
                            for m in abs_n_p+2:2:N
                                m2 = m^2
                                acoef = acoef + m * (m2 - (n - p)^2) * Vb[m]
                            end
                            A_χφ[n,abs_p] = A_χφ[n,abs_p] + complex(0.0, 0.5 * α * Re * c[abs_p] * acoef)
                        end
                    end
                    
                    for p in n+1:2:N
                        A_χμ[n,p] = 4.0 * Gf * p
                    end

                    # In χ rows, lambda coefficients
                    B_χχ[n,n] = complex(0.0, - G2 * α * Re * c[n])
                
                    # In μ rows, a_{φ,n}, a_{χ,n}, and a_{μ,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_μμ[n,p] = p * (p2 - n2)
                    end
                    
                    A_μμ[n,n] = - G2 * c[n] * α2
                    
                    for p in -N:N
                        abs_n_p = abs(n - p)
                        abs_p = abs(p)
                        if abs_n_p <= N
                            A_μμ[n,abs_p] = A_μμ[n,abs_p] + complex(0.0, - 0.5 * G2 * α * Re * c[abs_p] * c[abs_n_p] * Vb[abs_n_p])
                        end
                        
                        if abs_n_p+1 <= N
                            acoef = 0.0
                            for m in abs_n_p+1:2:N
                                acoef = acoef + m * Ub[m]
                            end
                            A_μφ[n,abs_p] = A_μφ[n,abs_p] + complex(0.0, Gf * α * Re * c[abs_p] * acoef)
                        end
                    end
                    
                    for p in n+1:2:N
                        A_μφ[n,p] = - 4.0 * Gf * p
                    end

                    # In μ rows, lambda coefficients
                    B_μμ[n,n] = complex(0.0, - G2 * α * Re * c[n])
                
                end
                
                # Set boundary conditions
                # For BC1 in φ
                for p in 0:2:N  # even mode
                    A_φφ[N-1,p] = 1.0
                end
                # For BC2 in φ
                for p in 1:2:N  # even mode
                    A_φφ[N,p] = 1.0
                end
                # For BC3 in φ (Z = -1)
                for p in 1:N
                    A_χφ[N-1,p] = ((-1.0)^p)*p^2
                end
                # For BC4 in χ (Z = 1)
                for p in 0:N
                    A_χχ[N,p] = 1.0
                end
                # For BC5 in μ (Z = -1)
                for p in 0:N
                    A_μμ[N-1,p] = (-1.0)^p
                end
                # For BC6 in μ (Z = 1)
                for p in 1:N
                    A_μμ[N,p] = p^2
                end
                
                # Gather A_φ, A_χ, A_μ, B_χ -> Acal, Bcal
                Acal[1:Ncal,1:Ncal] .= A_φφ[0:N,0:N]
                Acal[1:Ncal,Ncal+1:2*Ncal] .= A_φχ[0:N,0:N]
                Acal[Ncal+1:2*Ncal,1:Ncal] .= A_χφ[0:N,0:N]
                Acal[Ncal+1:2*Ncal,Ncal+1:2*Ncal] .= A_χχ[0:N,0:N]
                Acal[Ncal+1:2*Ncal,2*Ncal+1:Ntot] .= A_χμ[0:N,0:N]
                Acal[2*Ncal+1:Ntot,1:Ncal] .= A_μφ[0:N,0:N]
                Acal[2*Ncal+1:Ntot,Ncal+1:2*Ncal] .= A_μχ[0:N,0:N]
                Acal[2*Ncal+1:Ntot,2*Ncal+1:Ntot] .= A_μμ[0:N,0:N]
                Bcal[Ncal+1:2*Ncal,Ncal+1:2*Ncal] .= B_χχ[0:N,0:N]
                Bcal[2*Ncal+1:Ntot,2*Ncal+1:Ntot] .= B_μμ[0:N,0:N]
                
                # Convert Transpose type to Matrix (Array) type (ドット演算して要素だけコピーしている)
                A_J .= transpose(Acal)  # A_J = Transpose type
                B_J .= transpose(Bcal)  # B_J = Transpose type
                
                F = schur(A_J,B_J)
                Dcal = F.α ./ F.β
                D_J = imag.(Dcal)
                #for i in 1:Ntot
                #    println("Schur(Eig): ", Dcal[i])
                #end

                D_J[isnan.(D_J)] .= - Inf  # NaN -> - Inf
                nc_max = findmax(D_J)[2]
                c_out[i,j,k] = Dcal[nc_max]

            end
        end
    end

#println(D_J[findmax(F_J)[2]])
#println(F_J)
#println(V_infl)

    return c_out

end  # function full_EQ_set
    
function OSE_set( α_in, ε_in, Re_in, z_top, N_in )
# Solve the approximation (OSE) set of the equations (2.25), (2.27)-(2.28) in Lilly (1966)
    #-- arguments input:
    # α_in: wavenumber for y-direction [non-dim,Vector{Float64}]
    # ε_in: flow angle for Vg [degree,Vector{Float64}]
    # Re_in: Reinolds number [non-dim,Vector{Float64}]
    # Vb_in: Background flow spectra for V component [non-dim,Vector{Float64}]
    # z_top: Top height [non-dim,Vector{Float64}]
    # N_in: truncation wavenumber in vertical [1,Int64]
    #-- arguments output:
    # c_out: wave speed with the maximum in imaginary part [non-dim,Array{Complex{Float64},size(α),size(ε),size(Re)}]

    # set parameters
    d2r = π / 180.0  # degree to radian
    na = size(α_in)[1]
    ne = size(ε_in)[1]
    nr = size(Re_in)[1]
    N = N_in
    Ncal = N + 1
    Ntot = 2 * (N + 1)
    Gf = 0.5 * z_top
    G2 = Gf^2

    # initialize variables and matrix
    c_out = reshape(zeros(Complex{Float64},na,ne,nr),na,ne,nr)

    # initialize working arrays
    # A: non-lambda coefficients, B: lambda coefficients
    A_J = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    B_J = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    ### --- From here, supposing matrix element as row, col as in mathematical notation
    Acal = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    Bcal = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    A_φφ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for v in φ Chebyshev
    A_φχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in φ Chebyshev
    A_χφ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for φ in χ Chebyshev
    A_χχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in χ Chebyshev
    B_χχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in χ Chebyshev
    ### --- From here, supposing matrix element as col, row as in julia notation (only using these arrays for matrix calc)
    Vb = OffsetArray(zeros(Float64,Ncal),0:N)  # A vector composed of Chebyshev coefficients for the basic state flow of V component
    Vb_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of V component in the physical space
    c = OffsetArray(zeros(Float64,Ncal),0:N)  # factorized vector (c_0=2, c_n=1, (n>0))
    #-- For specifying the basic flows in physical space z_p (0 <= z_p <= z_top)
    θ = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in vertical coordinate (-π <= θ <= 0)
    Z = OffsetArray(zeros(Float64,Ncal),0:N)  # non-dimensional vertical coordinate (-1 <= Z <= 1)
    z_p = OffsetArray(zeros(Float64,Ncal),0:N)  # dimensional vertical coordinate (0 <= z_p <= z_top)

    # setting the factorized vector
    c[0] = 2.0
    c[1:N] .= 1.0

#    # setting basic flow spectra
#    Vb[0:N] .= Vb_in[1:Ncal]
    

    # set basic flow profiles in vertical
    θ[0:N] = [π * (j-N) / N for j in 0:N]  # non-dimensional vertical coordinate (-1<=Z<=1)
    Z[0:N] .= cos.(θ[0:N])
    z_p[0:N] .= Gf .* (Z[0:N] .+ 1.0)
    
    # setting maximum even and odd numbers from N
    Nemax = 0
    Nomax = 1
    for i in 0:2:N
        Nemax = i
    end
    for i in 1:2:N
        Nomax = i
    end
    #println("Nemax = ", Nemax, ", Nomax = ", Nomax)

    #-- Start loops for varying parameters (α, ε, Re)
    for k in 1:nr
        for j in 1:ne
            for i in 1:na
                α = α_in[i]
                ε = ε_in[j] * d2r
                Re = Re_in[k]
                α2 = α^2

                # initialize each matrices
                Acal = zeros(Complex{Float64},Ntot,Ntot)
                Bcal = zeros(Complex{Float64},Ntot,Ntot)
                A_φφ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_φχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χφ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                B_χχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]

                # set grid spacing and mean-wind profiles in vertical
                Vb_phys[0:N] .= -sin(ε) .+ exp.(-z_p[0:N]) .* sin.(z_p[0:N] .+ ε)
                #Vb_phys[0:N] .= 1.0 .- Z[0:N] .^2  # For the same flow in Orszag (1971)

                Vb = basic_flow_P2T( N, z_top, Vb_phys[0:N] )

                # setting matrices
                # 支配方程式から構築される部分 (チェビシェフ次数 0<=n<=N-2 の範囲)
                # この段階では境界条件で成り立つ係数の関係は考慮しない
                for n in 0:N-2
                    n2 = n^2
                    
                    # In v rows, a_{φ,n} and a_{χ,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_φφ[n,p] = p * (p2 - n2)
                    end
                    
                    A_φφ[n,n] = - G2 * c[n] * α2
                    A_φχ[n,n] = - G2 * c[n]
                    
                    # In χ rows, a_{φ,n} and a_{χ,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_χχ[n,p] = p * (p2 - n2)
                    end
                    
                    A_χχ[n,n] = - G2 * c[n] * α2
                    
                    for p in -N:N
                        abs_n_p = abs(n - p)
                        abs_p = abs(p)
                        if abs_n_p <= N
                            A_χχ[n,abs_p] = A_χχ[n,abs_p] + complex(0.0, - 0.5 * G2 * α * Re * c[abs_p] * c[abs_n_p] * Vb[abs_n_p])
                        end
                        
                        if abs_n_p+2 <= N
                            acoef = 0.0
                            for m in abs_n_p+2:2:N
                                m2 = m^2
                                acoef = acoef + m * (m2 - (n - p)^2) * Vb[m]
                            end
                            A_χφ[n,abs_p] = A_χφ[n,abs_p] + complex(0.0, 0.5 * α * Re * c[abs_p] * acoef)
                        end
                    end
                    
                    # In χ rows, lambda coefficients
                    B_χχ[n,n] = complex(0.0, - G2 * α * Re * c[n])
                
                end
                
                # Set boundary conditions
                # For BC1 in φ
                for p in 0:2:N  # even mode
                    A_φφ[N-1,p] = 1.0
#                    A_χφ[N-1,p] = p^2  # For the same boundary condition for Orszag OSE
                end
                # For BC2 in φ
                for p in 1:2:N  # even mode
                    A_φφ[N,p] = 1.0
#                    A_χφ[N,p] = p^2  # For the same boundary condition for Orszag OSE
                end
                # For BC3 in φ
                for p in 1:N
                    A_χφ[N-1,p] = ((-1.0)^p)*p^2
                end
                # For BC4 in χ
                for p in 0:N
                    A_χχ[N,p] = 1.0
                end
                
                # Gather A_φ, A_χ, A_μ, B_χ -> Acal, Bcal
                Acal[1:Ncal,1:Ncal] .= A_φφ[0:N,0:N]
                Acal[1:Ncal,Ncal+1:Ntot] .= A_φχ[0:N,0:N]
                Acal[Ncal+1:Ntot,1:Ncal] .= A_χφ[0:N,0:N]
                Acal[Ncal+1:Ntot,Ncal+1:Ntot] .= A_χχ[0:N,0:N]
                Bcal[Ncal+1:Ntot,Ncal+1:Ntot] .= B_χχ[0:N,0:N]
                
                # Convert Transpose type to Matrix (Array) type (ドット演算して要素だけコピーしている)
                A_J .= transpose(Acal)  # A_J = Transpose type
                B_J .= transpose(Bcal)  # B_J = Transpose type
                
                F = schur(A_J,B_J)
                Dcal = F.α ./ F.β
                D_J = imag.(Dcal)
                #for i in 1:Ntot
                #    println("Schur(Eig): ", Dcal[i])
                #end

                D_J[isnan.(D_J)] .= - Inf  # NaN -> - Inf
                nc_max = findmax(D_J)[2]
                c_out[i,j,k] = Dcal[nc_max]

            end
        end
    end

#println(D_J[findmax(F_J)[2]])
#println(F_J)
#println(V_infl)

    return c_out

end  # function OSE_set
    
function basic_flow_P2T( N_in, z_top, Vb_p )
    # Convert the basic flow in physical space to Chebyshev spectral space.
    # Following the conversion method in Ishioka (2004)
    #-- arguments input:
    # N_in: truncation wavenumber in vertical [1,Int64]
    # z_top: Top height [non-dim,Vector{Float64}]
    # Vb_p[1:N_in+1]: Basic flow in physical space of z_p (0 <= z_p <= z_top) [non-dim,Vector{Float64}]
    #-- arguments output:
    # Vb_sp: Basic flow in Chebyshev spectral space [non-dim,Vector{Float64}]

    N = N_in
    Ncal = N + 1
    Gf = 0.5 * z_top

    Vb_inter = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
    Vb_sp = OffsetArray(zeros(Float64,Ncal),0:N)  # Chebyshev spectra for the basic state flow
    Vb_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow in physical space
    Vb_sp_inter = OffsetArray(zeros(Complex{Float64},Ncal),0:N)  # intermediate variable in FFT for the basic state flow
    cbar = OffsetArray(zeros(Float64,Ncal),0:N)  # weighting coefficient for FFT
    FFTcal = zeros(Float64,Ncal)

    cbar[1:N-1] .= 1.0
    cbar[0] = 2.0
    cbar[N] = 2.0

    Vb_phys[0:N] .= Vb_p[1:Ncal]

    #-- Vb_phys -> V_inter
    for j in 0:N
        Vb_inter[j] = Vb_phys[j] + Vb_phys[N-j] + 2.0 * sin(j*π/N) * (Vb_phys[j] - Vb_phys[N-j])
    end

    FFTcal[1:Ncal] .= Vb_inter[0:N]
    FFTcal_sp = rfft(FFTcal[1:N]) .* (1.0/N)  # real FFT <- 両端は重複しているので片方のみ FFT に含める
    Vb_sp_inter[0:size(FFTcal_sp)[1]-1] .= FFTcal_sp[1:size(FFTcal_sp)[1]]

    Vb_sp[0] = real(Vb_sp_inter[0])  # even mode
    for j in 0:N  # wavenumber-1 mode
        Vb_sp[1] = Vb_sp[1] + Vb_phys[j] * cos(j*π/N) / cbar[j]
    end

    Vb_sp[1] = (2.0 / N) * Vb_sp[1]

    for l in 1:div(N,2)-1
        Vb_sp[2*l] = real(Vb_sp_inter[l])  # even mode
        Vb_sp[2*l+1] = Vb_sp[2*l-1] + imag(Vb_sp_inter[l])  # odd mode
    end

    Vb_sp[N] = real(Vb_sp_inter[div(N,2)])
    Vb_sp[0:N] .= Vb_sp[0:N] ./ cbar[0:N]

    # For checking the consistency between the input (original physical space) and output (converted by the definition)
    #for m in 0:N
    #    for j in 0:N
    #        Vb_ret[j] = Vb_ret[j] + Vb_sp[m] * cos(π*m*j/N)
    #    end
    #end
#Vbs = fft(Vb[1:N])
#println(size(Vbs), ",", Vbs./N)
#println(size(Vb_sp), ",", Vb_sp)

    return Vb_sp
    
end  # function basic_flow_P2T

end  # module Lilly1966_solver
```

```julia
# 基本場のチェビシェフ展開整合性チェック (function basic_flow_P2T)
# チェビシェフ展開への FFT の利用は石岡 (2004) に従う.
using OffsetArrays
#using FFTW
using PyPlot

N = 100
Ncal = N + 1
z_top = 40.0
Gf = 0.5 * z_top
ε = 0.0

Vb_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of V component in the physical space
Ub_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of U component in the physical space
Vb_inter = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Ub_inter = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Vb_ret = OffsetArray(zeros(Float64,Ncal),0:N)  # temporary array
Ub_ret = OffsetArray(zeros(Float64,Ncal),0:N)  # temporary array
Vb_sp = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Ub_sp = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Vb_sp_inter = OffsetArray(zeros(Complex{Float64},Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Ub_sp_inter = OffsetArray(zeros(Complex{Float64},Ncal),0:N)  # intermediate variable in FFT for the basic state flow
θ = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in vertical coordinate (-π <= θ <= 0)
Z = OffsetArray(zeros(Float64,Ncal),0:N)  # non-dimensional vertical coordinate (-1 <= Z <= 1)
z_p = OffsetArray(zeros(Float64,Ncal),0:N)  # dimensional vertical coordinate (0 <= z_p <= z_top)
cbar = OffsetArray(zeros(Float64,Ncal),0:N)  # weighting coefficient for FFT

FFTcal = zeros(Float64,Ncal)
#Vb_sp_inter = zeros(Complex{Float64},Ncal)
x =zeros(Float64,Ncal)
y1 =zeros(Float64,Ncal)
y1p =zeros(Float64,Ncal)
y2 =zeros(Float64,Ncal)
y2p =zeros(Float64,Ncal)

cbar[1:N-1] .= 1.0
cbar[0] = 2.0
cbar[N] = 2.0

# set basic flow profiles in vertical
θ[0:N] = [π * (j-N) / N for j in 0:N]  # non-dimensional vertical coordinate (-1<=Z<=1)
Z[0:N] .= cos.(θ[0:N])
z_p[0:N] .= Gf .* (Z[0:N] .+ 1.0)
Ub_phys[0:N] .= cos(ε) .- exp.(-z_p[0:N]) .* cos.(z_p[0:N] .+ ε)
Vb_phys[0:N] .= -sin(ε) .+ exp.(-z_p[0:N]) .* sin.(z_p[0:N] .+ ε)
#Vb_phys[0:N] .= 1.0 .- Z[0:N].^2

#-- {Vb_phys,Ub_phys} -> {V_inter,U_inter}
for j in 0:N
    Vb_inter[j] = Vb_phys[j] + Vb_phys[N-j] + 2.0 * sin(j*π/N) * (Vb_phys[j] - Vb_phys[N-j])
    Ub_inter[j] = Ub_phys[j] + Ub_phys[N-j] + 2.0 * sin(j*π/N) * (Ub_phys[j] - Ub_phys[N-j])
end

FFTcal[1:Ncal] .= Vb_inter[0:N]
#FFTcal_sp = rfft(FFTcal[1:N]) .* (1.0/N)  # <- 両端は重複しているので片方のみ FFT に含める
#Vb_sp_inter[0:size(FFTcal_sp)[1]-1] .= FFTcal_sp[1:size(FFTcal_sp)[1]]

FFTcal[1:Ncal] .= Ub_inter[0:N]
#FFTcal_sp = rfft(FFTcal[1:N]) .* (1.0/N)  # <- 両端は重複しているので片方のみ FFT に含める
#Ub_sp_inter[0:size(FFTcal_sp)[1]-1] .= FFTcal_sp[1:size(FFTcal_sp)[1]]

Vb_sp[0] = real(Vb_sp_inter[0])  # even mode
Ub_sp[0] = real(Ub_sp_inter[0])  # even mode
for j in 0:N  # wavenumber-1 mode
    Vb_sp[1] = Vb_sp[1] + Vb_phys[j] * cos(j*π/N) / cbar[j]
    Ub_sp[1] = Ub_sp[1] + Ub_phys[j] * cos(j*π/N) / cbar[j]
end
Vb_sp[1] = (2.0 / N) * Vb_sp[1]
Ub_sp[1] = (2.0 / N) * Ub_sp[1]
for l in 1:div(N,2)-1
    Vb_sp[2*l] = real(Vb_sp_inter[l])  # even mode
    Ub_sp[2*l] = real(Ub_sp_inter[l])  # even mode
    Vb_sp[2*l+1] = Vb_sp[2*l-1] + imag(Vb_sp_inter[l])  # odd mode
    Ub_sp[2*l+1] = Ub_sp[2*l-1] + imag(Ub_sp_inter[l])  # odd mode
end
Vb_sp[N] = real(Vb_sp_inter[div(N,2)])
Ub_sp[N] = real(Ub_sp_inter[div(N,2)])

Vb_sp[0:N] .= Vb_sp[0:N] ./ cbar[0:N]
Ub_sp[0:N] .= Ub_sp[0:N] ./ cbar[0:N]

for m in 0:N
    for j in 0:N
        Vb_ret[j] = Vb_ret[j] + Vb_sp[m] * cos(π*m*j/N)
        Ub_ret[j] = Ub_ret[j] + Ub_sp[m] * cos(π*m*j/N)
    end
end
#Vbs = fft(Vb[1:N])
#println(size(Vbs), ",", Vbs./N)
#println(size(Vb_sp), ",", Vb_sp)
x[1:Ncal] .= z_p[0:N]
y1[1:Ncal] .= Vb_phys[0:N]
y1p[1:Ncal] .= Vb_ret[0:N]
y2[1:Ncal] .= Ub_phys[0:N]
y2p[1:Ncal] .= Ub_ret[0:N]
scatter(x,[y1,y2])
#scatter(x,[y1,y1p,y2,y2p])
```
