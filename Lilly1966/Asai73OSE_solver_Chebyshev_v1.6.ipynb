# Asai & Nakasuji (1973) の Ekman 層不安定固有値解析 (Chebyshev 多項式展開)
[Asai & Nakasuji (1973)](https://doi.org/10.2151/jmsj1965.51.1_29) による Ekman 層における Parallel instability の固有値解析 (3 変数の Orr-Sommerfeld 方程式) を Chebyshev 多項式展開で実施する.

ただし, 切断波数$N$が大きい場合, 4 階微分項の評価が $O(N^7)$となり, 固有値計算時の丸め誤差が無視できなくなる可能性がある ([Dongarra et al. 1996](https://doi.org/10.1016/S0168-9274(96)00049-9)). そのため, 中間変数$\chi $を導入して 2 階微分のみで計算する ([Hibino et al. 2012](https://doi.org/10.2151/jmsj.2012-210)). 

\begin{equation}
\left(\dfrac{d^2}{dz^2} -k^2 \right) W\equiv \chi , \label{eq:Asai1973-B1-02}
\end{equation}
となる中間変数$\chi $を導入する. まず, チェビシェフ展開するために, 微分方程式の定義域を$0\leq z'\leq z'_{\mathrm{top}}\rightarrow \infty $から, $-1\leq Z\leq 1$に座標変換する:
$$Z\equiv f(z')=G^{-1}\left(z'-G \right) ,\quad G\equiv \dfrac{z'_{\mathrm{top}}}{2} $$
すると, Asai & Nakasuji (1973) の (3.14)-(3.16), および (\ref{eq:Asai1973-B1-02}) 式は
\begin{align}
&G^{-2}\ddot{Z} -\left[k^2+R_e\left(\sigma +ik_x\overline{u} +ik_y\overline{\upsilon} \right) \right] Z+G^{-1}\left[iR_e\left(k_y\dot{\overline{u}} -k_x\dot{\overline{\upsilon}} \right) W+2\dot{W} \right] =0, \label{eq:Asai1973-B1-04} \\
&G^{-2}\ddot{W} -\left(k^2W+\chi \right) =0, \label{eq:Asai1973-B1-05} \\
&G^{-2}\ddot{\chi} -\left[k^2 \chi +R_e\left(\sigma +i\overline{u} k_x+i\overline{\upsilon} k_y \right) \chi \right] +iG^{-2}R_e\left(k_x\ddot{\overline{u}} +k_y\ddot{\overline{\upsilon}} \right) W-2G^{-1}\dot{Z} -k^2R_eR_i\mathit{\Theta} =0, \label{eq:Asai1973-B1-06} \\
&G^{-2}\ddot{\mathit{\Theta}} -\left[k^2+P_rR_e\left(\sigma +ik_x\overline{u} +ik_y\overline{\upsilon} \right) \right] \mathit{\Theta} -G^{-1}P_rR_e^dot{W} =0. \label{eq:Asai1973-B1-07} 
\end{align}
Orszag (1971) の (13) 式にしたがって, 各変数を
\begin{equation}
\left[\begin{array}{c}
Z(z') \\
\chi (z') \\
W (z') \\
\mathit{\Theta} (z') \\
\overline{u} (z') \\
\overline{\upsilon} (z') \\
\overline{\theta} (z') \\
\end{array} \right] =\sum^{\infty}_{n=0}{\left[\begin{array}{c}
\hat{Z} _n \\
\hat{\chi} _n \\
\hat{W} _n \\
\hat{\mathit{\Theta}} _n \\
\hat{U} _n \\
\hat{V} _n \\
\hat{\theta} _n \\
\end{array} \right] T_n(z')} , \label{eq:Asai1973-B1-08}
\end{equation}
と展開する. (\ref{eq:Asai1973-B1-04})-(\ref{eq:Asai1973-B1-07}) 式のチェビシェフ展開はそれぞれ
\begin{align}
&\sum^{\infty}_{n=0}{\left\{G^{-2}\hat{Z}^{(2)}_n-\left[k^2+R_e\sigma \right] \hat{Z}_n-\dfrac{iR_e}{2} \left[k_xe_{U,Z,n}+k_ye_{V,Z,n} \right] \right.} \nonumber \\
&\qquad \qquad \left. +\dfrac{iR_e}{2G} \left(k_yf^{(1)}_{U,W,n}-k_xf^{(1)}_{V,W,n} \right) +\dfrac{2}{G} \hat{W}^{(1)}_n \right\} T_n=0, \label{eq:Asai1973-B1-10} \\
&\sum^{\infty}_{n=0}{\left\{G^{-2}\hat{W}^{(2)}_n-\left(k^2 \hat{W}_n+\hat{\chi}_n \right) \right\} T_n} =0, \label{eq:Asai1973-B1-11} \\
&\sum^{\infty}_{n=0}{\left\{G^{-2}\hat{\chi}^{(2)}_n-\left[k^2+R_e\sigma \right] \hat{\chi}_n-\dfrac{iR_e}{2} \left[k_xe_{U,\chi ,n}+k_ye_{V,\chi ,n} \right] +\dfrac{iR_e}{2G^2} \left(k_xf^{(2)}_{U,W,n}+k_yf^{(2)}_{V,W,n} \right) \right.} \nonumber \\
&\qquad \qquad \left. -\dfrac{2}{G} \hat{Z}^{(1)}_n-k^2R_eR_i\hat{\mathit{\Theta}}_n \right\} T_n=0, \label{eq:Asai1973-B1-12} \\
&\sum^{\infty}_{n=0}{\left\{G^{-2}\hat{\mathit{\Theta}}^{(2)}_n-\left[k^2+P_rR_e\sigma \right] \hat{\mathit{\Theta}} _n-\dfrac{iR_e}{2} \left[k_xe_{U,\mathit{\Theta} ,n}+k_ye_{V,\mathit{\Theta} ,n} \right] -\dfrac{P_rR_e}{2G} f^{(1)}_{\theta ,W,n} \right\} T_n} =0. \label{eq:Asai1973-B1-13}
\end{align}
ただし, Orszag (1971) の付録 (A7) より, 非線形項は
\begin{equation}
\begin{split}
e_{A,B,n}\equiv &\; \dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{A}_mc_{|n-m|}\hat{B}_{|n-m|}}, \\
f^{(1)}_{A,B,n} \equiv &\; \dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{B}_mc_{|n-m|}\hat{A}^{(1)}_{|n-m|}} =\dfrac{2}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{B}_m\sum^{\infty}_{\substack{p=|n-m|+1\\ p+|n-m|\equiv 1\; (\mathrm{mod} \; 2)}}{p\hat{A}_p}} , \\
f^{(2)}_{A,B,n} \equiv &\; \dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{B}_mc_{|n-m|}\hat{A}^{(2)}_{|n-m|}} =\dfrac{1}{c_n} \sum^{\infty}_{m=-\infty}{c_m\hat{B}_m\sum^{\infty}_{\substack{p=|n-m|+2\\ p\equiv |n-m|\; (\mathrm{mod} \; 2)}}{p(p^2-(n-m)^2)\hat{A}_p}} .
\end{split}
 \label{eq:Asai1973-B1-14}
\end{equation}
境界条件について, Asai & Nakasuji (1973) の (3.17) 式を$Z=\pm 1$で成り立つようにすると,
\begin{equation}
\begin{split}
&\sum^{N}_{p=0}{(-1)^p\hat{Z}_p} =0,\quad \sum^{N}_{p=0}{(-1)^p\hat{W}_p} =0,\quad \sum^{N}_{p=1}{p^2(-1)^{p-1}\hat{W}_p} =0,\quad \sum^{N}_{p=0}{(-1)^p\hat{\mathit{\Theta}}_p} =0,\qquad (z'=-1), \\
&\sum^{N}_{p=1}{p^2(+1)^{p-1}\hat{Z}_p} =0,\quad \sum^{N}_{p=0}{(+1)^p\hat{W}_p} =0,\quad \sum^{N}_{p=0}{(+1)^p\hat{\chi}_p} =0,\quad \sum^{N}_{p=0}{(+1)^p\hat{\mathit{\Theta}}_p} =0, \qquad (z'=1). \\
\end{split}
\label{eq:Asai1973-B1-26}
\end{equation}
(\ref{eq:Asai1973-3.26}) の第 3 式は$z'=1$では$W=\dfrac{d^2W}{dz'^2} =0$なので, (\ref{eq:Asai1973-B1-02}) 式から$\chi =0$が成り立つことを利用している. 

(\ref{eq:Asai1973-B1-21})-(\ref{eq:Asai1973-B1-24}) 式は$T_n$が$0\leq n\leq N-2$のそれぞれについて$4(N-1)$個の独立な方程式となる. これに加えて境界条件が 8 つの追加の関係を各係数$\hat{Z}_n,\; \hat{W}_n,\; \hat{\chi}_n,\; \hat{\mathit{\Theta}}$について与えるので, $4(N+1)$個の未知変数について, $4(N+1)$個の方程式が組み立てられる. 今, (\ref{eq:Asai1973-B1-21})-(\ref{eq:Asai1973-B1-24}) 式および境界条件 (\ref{eq:Asai1973-B1-26}) 式から
\begin{equation}
A\textbf{x} =cB\textbf{x} , \label{eq:Asai1973-B1-27}
\end{equation}
\begin{equation}
\textbf{x} \equiv [\hat{W}_0,\cdots ,\hat{W}_N,\hat{Z}_0,\cdots ,\hat{Z}_N,\hat{\chi}_0,\cdots ,\hat{\chi}_N,\hat{\mathit{\Theta}}_0,\cdots ,\hat{\mathit{\Theta}}_N]^T \label{eq:Asai1973-B1-28}
\end{equation}
という形の行列方程式が構築できる. $\textbf{x} $は未知変数で構築される$4(N+1)$次元の列ベクトル, $A,\; B$は$4(N+1)$次正方行列である. ここで, $A,B$はそれぞれ
\begin{equation}
A\equiv \left(\begin{array}{cccc}
A_{\ref{eq:Asai1973-B1-22},\hat{W}} & O & A_{\ref{eq:Asai1973-B1-22},\hat{\chi}} & O \\
\mathrm{BC1}   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
\mathrm{BC2}   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
A_{\ref{eq:Asai1973-B1-21},\hat{W}} & A_{\ref{eq:Asai1973-B1-21},\hat{Z}} & O & O \\
\textbf{0}^T   & \mathrm{BC3}  & \textbf{0}^T & \textbf{0}^T \\
\textbf{0}^T   & \mathrm{BC4}  & \textbf{0}^T & \textbf{0}^T \\
A_{\ref{eq:Asai1973-B1-23},\hat{W}} & A_{\ref{eq:Asai1973-B1-23},\hat{Z}} & A_{\ref{eq:Asai1973-B1-23},\hat{\chi}} & A_{\ref{eq:Asai1973-B1-23},\hat{\mathit{\Theta}}} \\
\mathrm{BC5}   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
\textbf{0}^T   & \textbf{0}^T  & \mathrm{BC6} & \textbf{0}^T \\
A_{\ref{eq:Asai1973-B1-24},\hat{W}} & O & O & A_{\ref{eq:Asai1973-B1-24},\hat{\mathit{\Theta}}} \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \mathrm{BC7} \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \mathrm{BC8} \\
\end{array} \right) ,\quad B\equiv \left(\begin{array}{cccc}
O          &      O     &      O     &      O     \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
O          & B_{\ref{eq:Asai1973-B1-21},\hat{Z}} & O & O \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
O          & O & B_{\ref{eq:Asai1973-B1-23},\hat{\chi}} & O \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
O          & O & O & B_{\ref{eq:Asai1973-B1-24},\hat{\mathit{\Theta}}} \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
\textbf{0}^T   & \textbf{0}^T  & \textbf{0}^T & \textbf{0}^T \\
\end{array} \right) , \label{eq:Asai1973-B1-29}
\end{equation}
ただし, 区分行列$A_{j,k}, B_{j,k}$は支配方程式 ($j$) において, 変数$k$の係数を表す$N-1$行$N+1$列の行列, $O$は$N-1$行$N+1$列のゼロ行列, $\textbf{0}$は$N+1$次のゼロベクトルとする. BC1-BC8 は 8 つの境界条件から得られる$\hat{Z}_n,\; \hat{W}_n,\; \hat{\chi}_n,\; \hat{\mathit{\Theta}}_n$に対する関係式である. 

```julia
module Asai1973_solver

using OffsetArrays  # 配列要素番号を任意開始にするモジュール (行列計算時は 1 始まりの通常 matrix に copy する)
using LinearAlgebra
using FFTW

export full_EQ_set, basic_flow_P2T

function full_EQ_set( k_in, ε_in, Re_in, Ri_in, Pr_in, z_top, β, N_in )
# Solve the complete set of the equations (2.25)-(2.28) in Asai & Nakasuji (1973)
    #-- arguments input:
    # k_in: wavenumber for y-direction [non-dim,Vector{Float64}]
    # ε_in: angle [degree,Vector{Float64}]
    # Re_in: Reinolds number [non-dim,Vector{Float64}]
    # Ri_in: Richardson number [non-dim,Vector{Float64}]
    # Pr_in: Prandtl number [non-dim,Vector{Float64}]
    # z_top: Model top height [non-dim,Vector{Float64}]
    # β: Vertical gradient of potential temperature [non-dim,Float64]
    # N_in: truncation wavenumber in vertical [1,Int64]
    #-- arguments output:
    # c_out: wave speed with the maximum in imaginary part [non-dim,Array{Complex{Float64},size(kx),size(ky),size(Re)}]

    # set parameters
    d2r = π / 180.0  # degree to radian
    nk = size(k_in)[1]
    ne = size(ε_in)[1]
    nr = size(Re_in)[1]
    N = N_in
    Ncal = N + 1
    Ntot = 4 * (N + 1)
    Gf = 0.5 * z_top
    G2 = Gf^2
    G_inv = 1.0 / Gf
    G2_inv = 1.0 / G2

    # initialize variables and matrix
    c_out = reshape(zeros(Complex{Float64},nk,ne,nr),nk,ne,nr)

    # initialize working arrays
    # A: non-lambda coefficients, B: lambda coefficients
    A_J = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    B_J = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    ### --- From here, supposing matrix element as row, col as in mathematical notation
    Acal = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    Bcal = reshape(zeros(Complex{Float64},Ntot,Ntot),Ntot,Ntot)  # Matrix for calculation
    A_ZZ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for Z in Z Chebyshev
    A_ZW = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for W in Z Chebyshev
    A_WW = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for W in W Chebyshev
    A_Wχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in W Chebyshev
    A_χZ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for Z in χ Chebyshev
    A_χW = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for W in χ Chebyshev
    A_χχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in χ Chebyshev
    A_χΘ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for Θ in χ Chebyshev
    A_ΘW = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for W in Θ Chebyshev
    A_ΘΘ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for Θ in Θ Chebyshev
    B_ZZ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for Z in Z Chebyshev
    B_χχ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for χ in χ Chebyshev
    B_ΘΘ = OffsetArray(Acal[1:Ncal,1:Ncal],0:N,0:N)  # coefficient matrix for Θ in Θ Chebyshev
    ### --- From here, supposing matrix element as col, row as in julia notation (only using these arrays for matrix calc)
    Vb = OffsetArray(zeros(Float64,Ncal),0:N)  # A vector composed of Chebyshev coefficients for the basic state flow of V component
    Ub = OffsetArray(zeros(Float64,Ncal),0:N)  # A vector composed of Chebyshev coefficients for the basic state flow of V component
    θb = OffsetArray(zeros(Float64,Ncal),0:N)  # A vector composed of Chebyshev coefficients for the basic stratification
    Vb_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of V component in the physical space
    Ub_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of U component in the physical space
    θb_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic stratification in the physical space
    c = OffsetArray(zeros(Float64,Ncal),0:N)  # factorized vector (c_0=2, c_n=1, (n>0))
    #-- For specifying the basic flows in physical space z_p (0 <= z_p <= z_top)
    φ = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in vertical coordinate (-π <= φ <= 0)
    z_p = OffsetArray(zeros(Float64,Ncal),0:N)  # non-dimensional vertical coordinate (-1 <= z_p <= 1)
    z = OffsetArray(zeros(Float64,Ncal),0:N)  # dimensional vertical coordinate (0 <= z <= z_top)

    # setting the factorized vector
    c[0] = 2.0
    c[1:N] .= 1.0

    # set basic flow profiles in vertical
    φ[0:N] = [π * (j-N) / N for j in 0:N]  # non-dimensional vertical coordinate (-1<=Z<=1)
    z_p[0:N] .= cos.(φ[0:N])
    z[0:N] .= Gf .* (z_p[0:N] .+ 1.0)

    # set grid spacing and mean-wind profiles in vertical
    Ub_phys[0:N] .= 1.0 .- exp.(-z_p[0:N]) .* cos.(z_p[0:N])
    Vb_phys[0:N] .= exp.(-z_p[0:N]) .* sin.(z_p[0:N])
    θb_phys[0:N] .= 1.0 .- β .* (z[0:N] .- Gf)

    Vb = basic_flow_P2T( N, z_top, Vb_phys[0:N] )
    Ub = basic_flow_P2T( N, z_top, Ub_phys[0:N] )
    θb = basic_flow_P2T( N, z_top, θb_phys[0:N] )

    # setting maximum even and odd numbers from N
    Nemax = 0
    Nomax = 1
    for i in 0:2:N
        Nemax = i
    end
    for i in 1:2:N
        Nomax = i
    end
    #println("Nemax = ", Nemax, ", Nomax = ", Nomax)

    #-- Start loops for varying parameters (k, ε, Re)
    for k in 1:nr
        for j in 1:ne
            for i in 1:nk
                kx = k_in[i] * tan(- ε_in[j] * d2r)
                ky = k_in[i]
                println("check k", kx, ", ", ky, ", ", k_in[i])
                Re = Re_in[k]
                Ri = Ri_in[k]
                Pr = Pr_in[k]

                # initialize each matrices
                Acal = zeros(Complex{Float64},Ntot,Ntot)
                Bcal = zeros(Complex{Float64},Ntot,Ntot)
                A_ZZ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_ZW[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_WW[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_Wχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χZ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χW[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_χΘ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_ΘW[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                A_ΘΘ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                B_ZZ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                B_χχ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]
                B_ΘΘ[0:N,0:N] .= Acal[1:Ncal,1:Ncal]

                # setting matrices
                # 支配方程式から構築される部分 (チェビシェフ次数 0<=n<=N-2 の範囲)
                # この段階では境界条件で成り立つ係数の関係は考慮しない
                for n in 0:N-2
                    n2 = n^2
                    k2 = kx^2 + ky^2
                    
                    # In W rows, a_{W,n} and a_{χ,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_WW[n,p] = G2_inv * p * (p2 - n2)
                    end
                    
                    A_WW[n,n] = - c[n] * k2
                    A_Wχ[n,n] = - c[n]
                    
                    # In Z rows, a_{Z,n} and a_{W,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_ZZ[n,p] = G2_inv * p * (p2 - n2)
                    end
                    
                    A_ZZ[n,n] = - c[n] * k2
                    
                    for p in -N:N
                        abs_n_p = abs(n - p)
                        abs_p = abs(p)
                        if abs_n_p <= N
                            A_ZZ[n,abs_p] = A_ZZ[n,abs_p] + complex(0.0, - 0.5 * Re * c[abs_p] * c[abs_n_p] * (kx * Ub[abs_n_p] + ky * Vb[abs_n_p]))
                        end
                        
                        if abs_n_p+1 <= N
                            acoef = 0.0
                            for m in abs_n_p+1:2:N
                                acoef = acoef + m * (ky * Ub[m] - kx * Vb[m])
                            end
                            A_ZW[n,abs_p] = A_ZW[n,abs_p] + complex(0.0, G_inv * Re * c[abs_p] * acoef)
                        end
                    end
                    
                    for p in n+1:2:N
                        A_ZW[n,p] = A_ZW[n,p] + 4.0 * G_inv * p
                    end

                    # In Z rows, eigenvalue coefficients
                    B_ZZ[n,n] = complex(Re * c[n], 0.0)
                
                    # In χ rows, a_{Z,n}, a_{W,n}, a_{χ,n}, and a_{Θ,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_χχ[n,p] = G2_inv * p * (p2 - n2)
                    end
                    
                    A_χχ[n,n] = - c[n] * k2
                    
                    for p in -N:N
                        abs_n_p = abs(n - p)
                        abs_p = abs(p)
                        if abs_n_p <= N
                            A_χχ[n,abs_p] = A_χχ[n,abs_p] + complex(0.0, - 0.5 * Re * c[abs_p] * c[abs_n_p] * (kx * Ub[abs_n_p] + ky * Vb[abs_n_p]))
                        end
                        
                        if abs_n_p+2 <= N
                            acoef = 0.0
                            for m in abs_n_p+2:2:N
                                m2 = m^2
                                acoef = acoef + m * (m2 - (n - p)^2) * (kx * Ub[m] + ky * Vb[m])
                            end
                            A_χW[n,abs_p] = A_χW[n,abs_p] + complex(0.0, 0.5 * G2_inv * Re * c[abs_p] * acoef)
                        end
                    end
                    
                    for p in n+1:2:N
                        A_χZ[n,p] = - 4.0 * G_inv * p
                    end

                    A_χΘ[n,n] = - k2 * Re * Ri * c[n]

                    # In χ rows, lambda coefficients
                    B_χχ[n,n] = complex(Re * c[n], 0.0)
                
                    # In Θ rows, a_{W,n} and a_{Θ,n}
                    for p in n+2:2:N
                        p2 = p^2
                        A_ΘΘ[n,p] = G2_inv * p * (p2 - n2)
                    end
                    
                    A_ΘΘ[n,n] = - c[n] * k2
                    
                    for p in -N:N
                        abs_n_p = abs(n - p)
                        abs_p = abs(p)
                        if abs_n_p <= N
                            A_ΘΘ[n,abs_p] = A_ΘΘ[n,abs_p] + complex(0.0, - 0.5 * Re * c[abs_p] * c[abs_n_p] * (kx * Ub[abs_n_p] + ky * Vb[abs_n_p]))
                        end
                        
                        if abs_n_p+1 <= N
                            acoef = 0.0
                            for m in abs_n_p+1:2:N
                                acoef = acoef + m * θb[m]
                            end
                            A_ΘW[n,abs_p] = A_ΘW[n,abs_p] + complex(0.0, - G_inv * Pr * Re * c[abs_p] * acoef)
                        end
                    end
                    
                    # In Θ rows, eigenvalue coefficients
                    B_ΘΘ[n,n] = complex(Pr * Re * c[n], 0.0)
                
                end
                
                # Set boundary conditions
                # For BC1 in Z (z_p = -1)
                for p in 0:N
                    A_ZZ[N-1,p] = (-1.0)^p
                end
                # For BC2 in Z (z_p = 1)
                for p in 1:N
                    A_ZZ[N,p] = p^2
                end
                # For BC3 in W
                for p in 0:2:N  # even mode
                    A_WW[N-1,p] = 1.0
                end
                # For BC4 in W
                for p in 1:2:N  # odd mode
                    A_WW[N,p] = 1.0
                end
                # For BC5 in W (z_p = -1)
                for p in 1:N
                    A_χW[N-1,p] = ((-1.0)^(p-1)) * p^2
                end
                # For BC6 in χ (z_p = 1)
                for p in 0:N
                    A_χχ[N,p] = 1.0
                end
                # For BC7 in Θ
                for p in 0:2:N  # even mode
                    A_ΘΘ[N-1,p] = 1.0
                end
                # For BC8 in Θ
                for p in 1:2:N  # odd mode
                    A_ΘΘ[N,p] = 1.0
                end
                
                # Gather A_W, A_Z, A_χ, A_Θ, B_Z, B_χ, B_Θ -> Acal, Bcal
                Acal[       1:  Ncal,       1:  Ncal] .= A_WW[0:N,0:N]
                Acal[       1:  Ncal,2*Ncal+1:3*Ncal] .= A_Wχ[0:N,0:N]
                Acal[  Ncal+1:2*Ncal,       1:  Ncal] .= A_ZW[0:N,0:N]
                Acal[  Ncal+1:2*Ncal,  Ncal+1:2*Ncal] .= A_ZZ[0:N,0:N]
                Acal[2*Ncal+1:3*Ncal,       1:  Ncal] .= A_χW[0:N,0:N]
                Acal[2*Ncal+1:3*Ncal,  Ncal+1:2*Ncal] .= A_χZ[0:N,0:N]
                Acal[2*Ncal+1:3*Ncal,2*Ncal+1:3*Ncal] .= A_χχ[0:N,0:N]
                Acal[2*Ncal+1:3*Ncal,3*Ncal+1:  Ntot] .= A_χΘ[0:N,0:N]
                Acal[3*Ncal+1:  Ntot,       1:  Ncal] .= A_ΘW[0:N,0:N]
                Acal[3*Ncal+1:  Ntot,3*Ncal+1:  Ntot] .= A_ΘΘ[0:N,0:N]
                Bcal[  Ncal+1:2*Ncal,  Ncal+1:2*Ncal] .= B_ZZ[0:N,0:N]
                Bcal[2*Ncal+1:3*Ncal,2*Ncal+1:3*Ncal] .= B_χχ[0:N,0:N]
                Bcal[3*Ncal+1:  Ntot,3*Ncal+1:  Ntot] .= B_ΘΘ[0:N,0:N]
                
                # Convert Transpose type to Matrix (Array) type (ドット演算して要素だけコピーしている)
                A_J .= transpose(Acal)  # A_J = Transpose type
                B_J .= transpose(Bcal)  # B_J = Transpose type
                
                F = schur(A_J,B_J)
                Dcal = F.α ./ F.β
                D_J = real.(Dcal)
                #for i in 1:Ntot
                #    println("Schur(Eig): ", Dcal[i])
                #end

                D_J[isnan.(D_J)] .= - Inf  # NaN -> - Inf
                nc_max = findmax(D_J)[2]
                c_out[i,j,k] = Dcal[nc_max]

            end
        end
    end

#println(D_J[findmax(F_J)[2]])
#println(F_J)
#println(V_infl)

    return c_out

end  # function full_EQ_set
    
function basic_flow_P2T( N_in, z_top, Vb_p )
    # Convert the basic flow in physical space to Chebyshev spectral space.
    # Following the conversion method in Ishioka (2004)
    #-- arguments input:
    # N_in: truncation wavenumber in vertical [1,Int64]
    # z_top: Top height [non-dim,Vector{Float64}]
    # Vb_p[1:N_in+1]: Basic flow in physical space of z_p (0 <= z_p <= z_top) [non-dim,Vector{Float64}]
    #-- arguments output:
    # Vb_sp: Basic flow in Chebyshev spectral space [non-dim,Vector{Float64}]

    N = N_in
    Ncal = N + 1
    Gf = 0.5 * z_top

    Vb_inter = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
    Vb_sp = OffsetArray(zeros(Float64,Ncal),0:N)  # Chebyshev spectra for the basic state flow
    Vb_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow in physical space
    Vb_sp_inter = OffsetArray(zeros(Complex{Float64},Ncal),0:N)  # intermediate variable in FFT for the basic state flow
    cbar = OffsetArray(zeros(Float64,Ncal),0:N)  # weighting coefficient for FFT
    FFTcal = zeros(Float64,Ncal)

    cbar[1:N-1] .= 1.0
    cbar[0] = 2.0
    cbar[N] = 2.0

    Vb_phys[0:N] .= Vb_p[1:Ncal]

    #-- Vb_phys -> V_inter
    for j in 0:N
        Vb_inter[j] = Vb_phys[j] + Vb_phys[N-j] + 2.0 * sin(j*π/N) * (Vb_phys[j] - Vb_phys[N-j])
    end

    FFTcal[1:Ncal] .= Vb_inter[0:N]
    FFTcal_sp = rfft(FFTcal[1:N]) .* (1.0/N)  # real FFT <- 両端は重複しているので片方のみ FFT に含める
    Vb_sp_inter[0:size(FFTcal_sp)[1]-1] .= FFTcal_sp[1:size(FFTcal_sp)[1]]

    Vb_sp[0] = real(Vb_sp_inter[0])  # even mode
    for j in 0:N  # wavenumber-1 mode
        Vb_sp[1] = Vb_sp[1] + Vb_phys[j] * cos(j*π/N) / cbar[j]
    end

    Vb_sp[1] = (2.0 / N) * Vb_sp[1]

    for l in 1:div(N,2)-1
        Vb_sp[2*l] = real(Vb_sp_inter[l])  # even mode
        Vb_sp[2*l+1] = Vb_sp[2*l-1] + imag(Vb_sp_inter[l])  # odd mode
    end

    Vb_sp[N] = real(Vb_sp_inter[div(N,2)])
    Vb_sp[0:N] .= Vb_sp[0:N] ./ cbar[0:N]

    # For checking the consistency between the input (original physical space) and output (converted by the definition)
    #for m in 0:N
    #    for j in 0:N
    #        Vb_ret[j] = Vb_ret[j] + Vb_sp[m] * cos(π*m*j/N)
    #    end
    #end
#Vbs = fft(Vb[1:N])
#println(size(Vbs), ",", Vbs./N)
#println(size(Vb_sp), ",", Vb_sp)

    return Vb_sp
    
end  # function basic_flow_P2T

end  # module Lilly1966_solver
```

```julia
# 基本場のチェビシェフ展開整合性チェック (function basic_flow_P2T)
# チェビシェフ展開への FFT の利用は石岡 (2004) に従う.
using OffsetArrays
#using FFTW
using PyPlot

N = 100
Ncal = N + 1
z_top = 40.0
Gf = 0.5 * z_top
ε = 0.0

Vb_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of V component in the physical space
Ub_phys = OffsetArray(zeros(Float64,Ncal),0:N)  # the basic state flow of U component in the physical space
Vb_inter = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Ub_inter = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Vb_ret = OffsetArray(zeros(Float64,Ncal),0:N)  # temporary array
Ub_ret = OffsetArray(zeros(Float64,Ncal),0:N)  # temporary array
Vb_sp = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Ub_sp = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Vb_sp_inter = OffsetArray(zeros(Complex{Float64},Ncal),0:N)  # intermediate variable in FFT for the basic state flow
Ub_sp_inter = OffsetArray(zeros(Complex{Float64},Ncal),0:N)  # intermediate variable in FFT for the basic state flow
θ = OffsetArray(zeros(Float64,Ncal),0:N)  # intermediate variable in vertical coordinate (-π <= θ <= 0)
Z = OffsetArray(zeros(Float64,Ncal),0:N)  # non-dimensional vertical coordinate (-1 <= Z <= 1)
z_p = OffsetArray(zeros(Float64,Ncal),0:N)  # dimensional vertical coordinate (0 <= z_p <= z_top)
cbar = OffsetArray(zeros(Float64,Ncal),0:N)  # weighting coefficient for FFT

FFTcal = zeros(Float64,Ncal)
#Vb_sp_inter = zeros(Complex{Float64},Ncal)
x =zeros(Float64,Ncal)
y1 =zeros(Float64,Ncal)
y1p =zeros(Float64,Ncal)
y2 =zeros(Float64,Ncal)
y2p =zeros(Float64,Ncal)

cbar[1:N-1] .= 1.0
cbar[0] = 2.0
cbar[N] = 2.0

# set basic flow profiles in vertical
θ[0:N] = [π * (j-N) / N for j in 0:N]  # non-dimensional vertical coordinate (-1<=Z<=1)
Z[0:N] .= cos.(θ[0:N])
z_p[0:N] .= Gf .* (Z[0:N] .+ 1.0)
Ub_phys[0:N] .= cos(ε) .- exp.(-z_p[0:N]) .* cos.(z_p[0:N] .+ ε)
Vb_phys[0:N] .= -sin(ε) .+ exp.(-z_p[0:N]) .* sin.(z_p[0:N] .+ ε)
#Vb_phys[0:N] .= 1.0 .- Z[0:N].^2

#-- {Vb_phys,Ub_phys} -> {V_inter,U_inter}
for j in 0:N
    Vb_inter[j] = Vb_phys[j] + Vb_phys[N-j] + 2.0 * sin(j*π/N) * (Vb_phys[j] - Vb_phys[N-j])
    Ub_inter[j] = Ub_phys[j] + Ub_phys[N-j] + 2.0 * sin(j*π/N) * (Ub_phys[j] - Ub_phys[N-j])
end

FFTcal[1:Ncal] .= Vb_inter[0:N]
#FFTcal_sp = rfft(FFTcal[1:N]) .* (1.0/N)  # <- 両端は重複しているので片方のみ FFT に含める
#Vb_sp_inter[0:size(FFTcal_sp)[1]-1] .= FFTcal_sp[1:size(FFTcal_sp)[1]]

FFTcal[1:Ncal] .= Ub_inter[0:N]
#FFTcal_sp = rfft(FFTcal[1:N]) .* (1.0/N)  # <- 両端は重複しているので片方のみ FFT に含める
#Ub_sp_inter[0:size(FFTcal_sp)[1]-1] .= FFTcal_sp[1:size(FFTcal_sp)[1]]

Vb_sp[0] = real(Vb_sp_inter[0])  # even mode
Ub_sp[0] = real(Ub_sp_inter[0])  # even mode
for j in 0:N  # wavenumber-1 mode
    Vb_sp[1] = Vb_sp[1] + Vb_phys[j] * cos(j*π/N) / cbar[j]
    Ub_sp[1] = Ub_sp[1] + Ub_phys[j] * cos(j*π/N) / cbar[j]
end
Vb_sp[1] = (2.0 / N) * Vb_sp[1]
Ub_sp[1] = (2.0 / N) * Ub_sp[1]
for l in 1:div(N,2)-1
    Vb_sp[2*l] = real(Vb_sp_inter[l])  # even mode
    Ub_sp[2*l] = real(Ub_sp_inter[l])  # even mode
    Vb_sp[2*l+1] = Vb_sp[2*l-1] + imag(Vb_sp_inter[l])  # odd mode
    Ub_sp[2*l+1] = Ub_sp[2*l-1] + imag(Ub_sp_inter[l])  # odd mode
end
Vb_sp[N] = real(Vb_sp_inter[div(N,2)])
Ub_sp[N] = real(Ub_sp_inter[div(N,2)])

Vb_sp[0:N] .= Vb_sp[0:N] ./ cbar[0:N]
Ub_sp[0:N] .= Ub_sp[0:N] ./ cbar[0:N]

for m in 0:N
    for j in 0:N
        Vb_ret[j] = Vb_ret[j] + Vb_sp[m] * cos(π*m*j/N)
        Ub_ret[j] = Ub_ret[j] + Ub_sp[m] * cos(π*m*j/N)
    end
end
#Vbs = fft(Vb[1:N])
#println(size(Vbs), ",", Vbs./N)
#println(size(Vb_sp), ",", Vb_sp)
x[1:Ncal] .= z_p[0:N]
y1[1:Ncal] .= Vb_phys[0:N]
y1p[1:Ncal] .= Vb_ret[0:N]
y2[1:Ncal] .= Ub_phys[0:N]
y2p[1:Ncal] .= Ub_ret[0:N]
scatter(x,[y1,y2])
#scatter(x,[y1,y1p,y2,y2p])
```
